```python
import torch 
from torch import nn
import matplotlib.pyplot as plt
```

```python
# simple linear regression
weight = 0.7
bias = 0.3

start = 0
end = 1
step = 0.02
X = torch.arange(start, end, step).unsqueeze(dim=1)
y = weight * X + bias
len(X), len(y)
```

```python
train_split = int(0.8 * len(X))
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
def plot_pred(train_data=X_train, train_labels=y_train, test_data=X_test, test_labels=y_test, preds=None):
    plt.figure(figsize=(10, 7))
    
    plt.scatter(train_data, train_labels, c='b', s=4, label='Training data')
    plt.scatter(test_data, test_labels, c='g', s=4, label='Testing data')
    if preds is not None:
        plt.scatter(test_data, preds, c='r', s=4, label='Predictions')
    plt.legend(prop={"size": 14})
    
plot_pred()
```

```python
class LRModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.weights = nn.Parameter(torch.randn(1,
                                               requires_grad=True,
                                               dtype=torch.float))
        self.bias = nn.Parameter(torch.randn(1,
                                            requires_grad=True,
                                            dtype=torch.float))
        
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.weights * x + self.bias
```

```python
model_0 = LRModel()
print(list(model_0.parameters()))
print(model_0.state_dict())
```

```python
with torch.inference_mode(): # turns off the gradient tracking for calculations; use it instead of torch.no_grad()
    y_preds = model_0(X_test)
print(y_preds)
plot_pred(preds=y_preds)
```

```python
loss_fn = nn.L1Loss() # loss function: calculates loss
optim = torch.optim.SGD(params=model_0.parameters(), lr=0.01) # optimizer: changes parameters to reduce loss
```

```python
torch.manual_seed(42)

epochs = 1000

epoch_count = []
loss_values = []
test_loss_values = []

print('Target: ', weight, bias)
print(model_0.state_dict())
for epoch in range(1, epochs+1):

    model_0.train() # turns on gradient tracking
    train_pred = model_0(X_train) # prediction
    loss = loss_fn(train_pred, y_train) # calculates loss value
    optim.zero_grad() # sets gradients to 0
    loss.backward() # backpropagation
    optim.step() # gradient descent
    
    model_0.eval() # turns off gradient tracking
    with torch.no_grad():
        test_pred = model_0(X_test)
        test_loss = loss_fn(test_pred, y_test)
        
    if epoch % 20 == 0:
        epoch_count.append(epoch)
        loss_values.append(loss)
        test_loss_values.append(test_loss)
        print(f"Epoch: {epoch} Loss: {loss} Test Loss: {test_loss}")    

print(model_0.state_dict())
```

```python
with torch.inference_mode(): # turns off the gradient tracking for calculations; use it instead of torch.no_grad()
    y_preds = model_0(X_test)
print(y_preds)
plot_pred(preds=y_preds)
```

```python
import numpy as np

plt.plot(epoch_count, torch.tensor(loss_values).cpu().numpy(), label='Train Loss')
plt.plot(epoch_count, test_loss_values, label='Test Loss')
plt.title('Training and test loss curves')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
```

```python
# Optimizer has also its own state_dict like model has saved state dictionary
# model.save() saves the entire model and model.load() loads the entire model, 
# whereas model.save_state_dict() saves only learnable parameters, while model.load_state_dict() loads the learnable parameters
from pathlib import Path

# Create models directory
MODEL_PATH = Path('model')
MODEL_PATH.mkdir(parents=True, exist_ok=True)
# Create model save path
MODEL_NAME = '01_pytorch_workflow_fundamentals_0.pth'
MODEL_SAVE_PATH = MODEL_PATH / MODEL_NAME
print(MODEL_SAVE_PATH)
# saving with state_dict is recommended
torch.save(obj=model_0.state_dict(), f=MODEL_SAVE_PATH)
# Load the model state dict
loaded_model_0 = LRModel()
print(loaded_model_0.state_dict())
loaded_model_0.load_state_dict(torch.load(MODEL_SAVE_PATH))
print(loaded_model_0.state_dict())
```

```python
loaded_model_0.eval()
with torch.inference_mode():
    pred = loaded_model_0(X_test)
print(pred)
```

Putting it all together

```python
import torch
from torch import nn
import matplotlib.pyplot as plt

device = 'cuda' if torch.cuda.is_available() else 'cpu'

## 1. Data preparation
weight = 0.9
bias = 0.6
start = 0
end = 1
step = 0.02
X = torch.arange(start, end, step).unsqueeze(dim=1)
y = weight * X + bias

train_split = int(0.8 * len(X))
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]
# plot_pred(X_train, y_train, X_test, y_test)

## 2. Build the model
class LRV2(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_layer = nn.Linear(in_features=1, out_features=1)
        
    def forward(self, x):
        return self.linear_layer(x)
    
torch.manual_seed(42)
model_1 = LRV2()
# print(model_1.state_dict())

## 3. Train and test the model
# print(next(model_1.parameters()).device) # checks at which device the model is
model_1.to(device)
X_train, y_train, X_test, y_test = X_train.to(device), y_train.to(device), X_test.to(device), y_test.to(device)
epochs = 1500
loss_fn = nn.L1Loss()
optim = torch.optim.SGD(params=model_1.parameters(), lr=0.001)
for epoch in range(1, epochs+1):
    model_1.train()
    
    pred = model_1(X_train)
    loss = loss_fn(pred, y_train)
    optim.zero_grad()
    loss.backward()
    optim.step()
    
    model_1.eval()
    with torch.inference_mode():
        test_pred = model_1(X_test)
        test_loss = loss_fn(test_pred, y_test)
#     if epoch % 10 == 0:
#         print('Loss:', loss, 'Test Loss:', test_loss)
# print(model_1.state_dict())

## 4. Inference
model_1.eval()
with torch.inference_mode():
    y_pred = model_1(X_test)
# plot_pred(X_train, y_train, X_test, y_test, preds=y_pred.cpu())

## 5. Save and load model
from pathlib import Path

MODEL_PATH = Path('models')
MODEL_PATH.mkdir(parents=True, exist_ok=True)
MODEL_NAME = "01_pytorch_workflow_fundamentals_1.pth"
MODEL_SAVE_PATH = MODEL_PATH / MODEL_NAME
# print(MODEL_SAVE_PATH)
torch.save(model_1.state_dict(), MODEL_SAVE_PATH)

loaded_model_1 = LRV2()
loaded_model_1.load_state_dict(torch.load(MODEL_SAVE_PATH))
loaded_model_1.to(device)
# print(loaded_model_1.state_dict())

loaded_model_1.eval()
with torch.inference_mode():
    preds = loaded_model_1(X_test)
# y_pred == preds
```
